```
def lomuto_partition(arr, low, high):
    pivot = arr[high]        # take last element as pivot
    i = low                  # index of smaller element

    for j in range(low, high):
        print(arr[j],pivot)  # loop from low to high-1
        if arr[j] <= pivot:
            print("hdf",arr[j],pivot)
            print("---",arr[j],arr[i])   # if current element <= pivot
            arr[i], arr[j] = arr[j], arr[i]  # swap
            i += 1               # move boundary of smaller elements
    print(arr)
    # finally place pivot in correct position
    arr[i], arr[high] = arr[high], arr[i]
    print(arr)
    return i   # return pivot index


def quick_sort(arr, low, high):
    if low < high:
        p = lomuto_partition(arr, low, high)  # partition index
        quick_sort(arr, low, p - 1)           # sort left side
        quick_sort(arr, p + 1, high)          # sort right side


# Example
arr = [10, 80, 30, 90, 40, 50, 70]
print("Before sorting:", arr)
quick_sort(arr, 0, len(arr) - 1)
print("After sorting:", arr)
```

```
#brooforce approach for leetcode 146 lru cache
class LRUCache(object):

    def __init__(self, capacity):
        self.d = {}
        self.l = []
        self.cap = capacity

    def get(self, key):
        if key in self.d:
            self.l.remove(key)
            self.l.append(key)
            return self.d[key]
        return -1

    def put(self, key, value):
        if key in self.d:
            self.d[key] = value
            self.l.remove(key)
            self.l.append(key)
        else:
            if len(self.d) >= self.cap:
                lru = self.l.pop(0)
                self.d.pop(lru)
            self.d[key] = value
            self.l.append(key)
```

```
class TrieNode:
    def __init__(self):
        self.children = {}
        self.is_end = False


class Trie:

    def __init__(self):
        self.root = TrieNode()

    def insert(self, word):
        node = self.root
        for ch in word:
            if ch not in node.children:
                node.children[ch] = TrieNode()
            node = node.children[ch]
        node.is_end = True

    def search(self, word):
        node = self.root
        for ch in word:
            if ch not in node.children:
                return False
            node = node.children[ch]
        return node.is_end

    def startsWith(self, prefix):
        node = self.root
        for ch in prefix:
            if ch not in node.children:
                return False
            node = node.children[ch]
        return True
```

```
#1317. Convert Integer to the Sum of Two No-Zero Integers
#No-Zero integer is a positive integer that does not contain any 0 in its decimal representation.
class Solution(object):
    def getNoZeroIntegers(self, n):
        """
        :type n: int
        :rtype: List[int]
        """
        def checkNoZero(num):
            while num > 0:
                remain = num % 10
                if remain == 0:
                    return False
                # chop off last digit
                num = num // 10
            return True
        a = 1
        b = n -1
        while not checkNoZero(a) or not checkNoZero(b):
            a += 1
            b -= 1
        return [a, b]
```

```
import heapq

class Solution(object):
    def mergeKLists(self, lists):
        heap = []

        # only push the head of each list
        for node in lists:
            if node:
                heapq.heappush(heap, (node.val, id(node), node))

        head = ListNode()
        prev = head

        while heap:
            _, _, node = heapq.heappop(heap)
            prev.next = node
            prev = prev.next
            if node.next:
                heapq.heappush(heap, (node.next.val, id(node.next), node.next))

        return head.next
```

```
from collections import deque

def preorderToLevelOrder(preorder):
    if not preorder:
        return []

    result = []
    q = deque([preorder])   # process subarrays like we did for inorder

    while q:
        sub = q.popleft()
        if not sub:
            continue

        root = sub[0]   # root is always first element in preorder
        result.append(root)

        # split into left and right parts
        left = [x for x in sub[1:] if x < root]
        right = [x for x in sub[1:] if x > root]

        if left:
            q.append(left)
        if right:
            q.append(right)

    return result

# Example
preorder = [4, 2, 1, 3, 6, 5, 7]
print(preorderToLevelOrder(preorder))
```
